```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
total=purchase_data["SN"].nunique()
total_df=pd.DataFrame({'Total Players':[purchase_data["SN"].nunique()]})
total_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
item=purchase_data["Item ID"].nunique()
average=purchase_data["Price"].mean()
purchase=purchase_data["Purchase ID"].count()
revenue=purchase_data["Price"].sum()
summary1_df=pd.DataFrame({
    "Number of Unique Items":[item],
    "Average Price":[average],
    "Number of Purchases":[purchase],
    "Total Revenue":[revenue]
})
average=pd.options.display.float_format="${0:,.2f}".format
revenue=pd.options.display.float_format="${0:,.2f}".format
summary1_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
male_df=purchase_data.loc[purchase_data["Gender"]=="Male",:]
male_count=male_df["SN"].nunique()
male_percent=male_count/total*100
female_df=purchase_data.loc[purchase_data["Gender"]=="Female",:]
female_count=female_df["SN"].nunique()
female_percent=female_count/total*100
other=total-female_count-male_count
other_percent=other/total*100
summary2_df=pd.DataFrame({
    "":["Male","Female","Other / Non-Disclosed"],
    "Total Count":[male_count,female_count,other],
    "Percentage of Players":[male_percent,female_percent,other_percent]
})
pd.options.display.float_format="{0:,.2f}%".format
summary2_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# filter by user id first
filtered_df=purchase_data.groupby(["SN"])
filtered_total_df=filtered_df["Price"].sum()
filtered_itemcount_df=filtered_df["Item Name"].size()
filtered_gender_df=filtered_df["Gender"].first()
data_sn_df=pd.merge(filtered_itemcount_df,filtered_total_df,on="SN",how="outer")
data_by_sn_df=pd.merge(data_sn_df,filtered_gender_df,on="SN",how="outer")
```

```python
male_df=data_by_sn_df.loc[data_by_sn_df["Gender"]=="Male"]
male_itemcount=male_df["Item Name"].sum()
male_totalpurchase=male_df["Price"].sum()
male_avgprice=male_totalpurchase/male_itemcount
male_pp=male_totalpurchase/male_count
female_df=data_by_sn_df.loc[data_by_sn_df["Gender"]=="Female"]
female_itemcount=female_df["Item Name"].sum()
female_totalpurchase=female_df["Price"].sum()
female_avgprice=female_totalpurchase/female_itemcount
female_pp=female_totalpurchase/female_count
other_df=data_by_sn_df.loc[data_by_sn_df["Gender"]=="Other / Non-Disclosed"]
other_itemcount=other_df["Item Name"].sum()
other_totalpurchase=other_df["Price"].sum()
other_avgprice=other_totalpurchase/other_itemcount
other_pp=other_totalpurchase/other
summary3_df=pd.DataFrame({
    "Gender":["Male","Female","Other / Non-Disclosed"],
    "Purchase Count":[male_itemcount,female_itemcount,other_itemcount],
    "Average Purchase Price":[male_avgprice,female_avgprice,other_avgprice],
    "Total Purchase Value":[male_totalpurchase,female_totalpurchase,other_totalpurchase],
    "Avg Total Purchase per Person":[male_pp,female_pp,other_pp]
})
pd.options.display.float_format="${0:,.2f}".format
summary3_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins=[0,9,14,19,24,29,34,39,150]
age_label=["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
purchase_data["Age Range"]=pd.cut(purchase_data["Age"],bins,labels=age_label)
binned_df=purchase_data.groupby(["Age Range"])
age_count_df=binned_df["SN"].nunique()
age_percent_df=age_count_df/total*100
summary4_df=pd.merge(age_count_df,age_percent_df,on="Age Range",how="outer")
summary4_df=summary4_df.rename(
    columns={
        "SN_x":"Total Count",
        "SN_y":"Percentage of Players"
})
pd.options.display.float_format="{0:,.2f}%".format
summary4_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
age_purchasecount_df=binned_df["Item ID"].count()
age_purchasetotal_df=binned_df["Price"].sum()
age_avgprice_df=age_purchasetotal_df/age_purchasecount_df
age_pp_df=age_purchasetotal_df/age_count_df
summary5_df=pd.concat([age_purchasecount_df,age_avgprice_df,age_purchasetotal_df,age_pp_df],axis=1)
summary5_df.columns=['Purchase Count', 'Avg Total Purchase per Person', 'Total Purchase Value', "Average Purchase Price"]
pd.options.display.float_format="${0:,.2f}".format
summary5_df
```
